```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('butterfly.jpeg')

image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Data preparation

image_copy = np.copy(image)

pixel_vals = image_copy.reshape((-1,3))
pixel_vals = np.float32(pixel_vals)
```

```python
# K-MEANS

k = 3

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER , 10 , 1.0)

retval,labels,centers = cv2.kmeans(pixel_vals,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
# display

centers = np.uint8(centers)

segmented_data = centers[labels.flatten()]

segmented_image = segmented_data.reshape((image_copy.shape))

labels_reshape = labels.reshape(image_copy.shape[0] , image_copy.shape[1])

plt.imshow(segmented_image)
```

```python
fig , (ax1,ax2,ax3) = plt.subplot(1,3,figsize=(10,10))

ax1.imshow(labels_reshape == 0)
```
